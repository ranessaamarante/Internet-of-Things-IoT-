# Parte 2


O conjunto de dados escolhido foi o "Breast Cancer Wisconsin (Diagnostic)" do scikit-learn, que contém características computadas a partir de imagens digitalizadas de aspirados de agulha fina de uma massa mamária. O objetivo é prever se o tumor identificado é benigno (B) ou maligno (M).

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Carregando o conjunto de dados "Breast Cancer Wisconsin (Diagnostic)"
breast_cancer = load_breast_cancer()

# Dividindo o conjunto de dados em dados de treinamento e teste (80% treinamento, 20% teste)
X_train, X_test, y_train, y_test = train_test_split(breast_cancer.data, breast_cancer.target, test_size=0.2, random_state=42)

```

# SVM

```python
# Inicializando e treinando o modelo SVM
svm_model = SVC()
svm_model.fit(X_train, y_train)

# Fazendo previsões no conjunto de teste e calculando a acurácia
svm_predictions = svm_model.predict(X_test)
svm_accuracy = accuracy_score(y_test, svm_predictions)

# Exibindo a acurácia, o relatório das classificações e a matriz de confusão do modelo 
print("Acurácia do SVM:", svm_accuracy)
print("Relatório de Classificação do SVM:\n", classification_report(y_test, svm_predictions))
```

```python
# Plotando a matriz de confusão colorida para o modelo SVM
svm_cm = confusion_matrix(y_test, svm_predictions)
plt.figure(figsize=(5, 5))
sns.heatmap(svm_cm, annot=True, cmap='Blues', fmt='g', cbar=False)
plt.xlabel('Rótulos Previstos')
plt.ylabel('Rótulos Verdadeiros')
plt.title('Matriz de Confusão - SVM')
plt.savefig('p2_cm_svm.svg', format='svg')
plt.show()
```

# KNN

```python
# Repetindo o processo para o modelo KNN
knn_model = KNeighborsClassifier()
knn_model.fit(X_train, y_train)
knn_predictions = knn_model.predict(X_test)
knn_accuracy = accuracy_score(y_test, knn_predictions)

# Exibindo a acurácia, o relatório das classificações e a matriz de confusão do modelo
print("\nAcurácia do KNN:", knn_accuracy)
print("Relatório de Classificação do KNN:\n", classification_report(y_test, knn_predictions))
```

```python
knn_cm = confusion_matrix(y_test, knn_predictions)
plt.figure(figsize=(5, 5))
sns.heatmap(knn_cm, annot=True, cmap='Greens', fmt='g', cbar=False)
plt.xlabel('Rótulos Previstos')
plt.ylabel('Rótulos Verdadeiros')
plt.title('Matriz de Confusão - KNN')
plt.savefig('p2_cm_knn.svg', format='svg')
plt.show()
```

# Decision Tree

```python
# Repetindo o processo para o modelo Decision Tree
dt_model = DecisionTreeClassifier()
dt_model.fit(X_train, y_train)
dt_predictions = dt_model.predict(X_test)
dt_accuracy = accuracy_score(y_test, dt_predictions)

# Exibindo a acurácia, o relatório das classificações e a matriz de confusão do modelo 
print("\nAcurácia do Decision Tree:", dt_accuracy)
print("Relatório de Classificação do Decision Tree:\n", classification_report(y_test, dt_predictions))
```

```python
dt_cm = confusion_matrix(y_test, dt_predictions)
plt.figure(figsize=(5, 5))
sns.heatmap(dt_cm, annot=True, cmap='Oranges', fmt='g', cbar=False)
plt.xlabel('Rótulos Previstos')
plt.ylabel('Rótulos Verdadeiros')
plt.title('Matriz de Confusão - Decision Tree')
plt.savefig('p2_cm_dt.svg', format='svg')
plt.show()
```

# Random Forest

```python
# Repetindo o processo para o modelo Random Forest
rf_model = RandomForestClassifier()
rf_model.fit(X_train, y_train)
rf_predictions = rf_model.predict(X_test)
rf_accuracy = accuracy_score(y_test, rf_predictions)

# Exibindo a acurácia, o relatório das classificações e a matriz de confusão do modelo 
print("\nAcurácia do Random Forest:", rf_accuracy)
print("Relatório de Classificação do Random Forest:\n", classification_report(y_test, rf_predictions))
```

```python
rf_cm = confusion_matrix(y_test, rf_predictions)
plt.figure(figsize=(5, 5))
sns.heatmap(rf_cm, annot=True, cmap='BuPu', fmt='g', cbar=False)
plt.xlabel('Rótulos Previstos')
plt.ylabel('Rótulos Verdadeiros')
plt.title('Matriz de Confusão - Random Forest')
plt.savefig('p2_cm_rf.svg', format='svg')
plt.show()
```

# Naive Bayes

```python
# Repetindo o processo para o modelo Naive Bayes (Gaussian)
nb_model = GaussianNB()
nb_model.fit(X_train, y_train)
nb_predictions = nb_model.predict(X_test)
nb_accuracy = accuracy_score(y_test, nb_predictions)

# Exibindo a acurácia, o relatório das classificações e a matriz de confusão do modelo
print("\n Acurácia do Naive Bayes:", nb_accuracy)
print("Relatório de Classificação do Naive Bayes:\n", classification_report(y_test, nb_predictions))
```

```python
nb_cm = confusion_matrix(y_test, nb_predictions)
plt.figure(figsize=(5, 5))
sns.heatmap(nb_cm, annot=True, cmap='Purples', fmt='g', cbar=False)
plt.xlabel('Rótulos Previstos')
plt.ylabel('Rótulos Verdadeiros')
plt.title('Matriz de Confusão - Naive Bayes (Gaussian)')
plt.savefig('p2_cm_naive.svg', format='svg')
plt.show()
```
